```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df.head(-1)
```

```python
def EIIP(sequence):
    sequence = list(sequence)
    for i in range(len(sequence)):
        if sequence[i] == 'G':
            sequence[i] = 0.0806
        elif sequence[i] == 'A':
            sequence[i] = 0.1260
        elif sequence[i] == 'T':
            sequence[i] = 0.1335
        elif sequence[i] == 'C':
            sequence[i] = 0.1340
    return sequence
```

```python
df['Sequence'] = df['Sequence'].map(lambda x: EIIP(x))
```

```python
df
```

```python
for i in range(len(df)):
    for j in range(len(df['Sequence'][i])):
        if str(df['Sequence'][i][j]).isalpha():
            df['Sequence'][i][j] = 0
```

```python
from scipy.signal import stft
from scipy.stats import skew, kurtosis
```

```python
def convertMappingToSTFT(array, label):
    f, t, zxx = stft(array)

    # flatten the array
    flat = zxx.flatten()
    real = [ele.real for ele in flat]
    imag = [ele.imag for ele in flat]

    # calculate features
    # 1. sum
    # 2. mean
    # 3. variance
    # 4. standard deviation
    # 5. skewness
    # 6. kurtosis
    data = [np.sum(real), np.sum(imag), np.mean(real), np.mean(imag), np.std(real), np.std(imag), np.var(real),
            np.var(imag), skew(real), skew(imag), kurtosis(real), kurtosis(imag), label]
    return data
```

```python
arr = []
for i in range(len(df)):
    arr.append(convertMappingToSTFT(np.array(df['Sequence'][i]).astype(np.float64), df['label'][i]))
```

```python
features = pd.DataFrame(arr, columns=['r_sum', 'i_sum', 'r_mean', 'i_mean', 'r_std', 'i_std', 'r_var', 'i_var', 'r_skew', 'i_skew', 'r_kurtosis', 'i_kurtosis', 'label'])
```
